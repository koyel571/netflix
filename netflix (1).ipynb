```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python

data=pd.read_csv('/kaggle/input/netflix-data-cleaning-analysis-and-visualization/netflix1.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data1=data.drop(["show_id","date_added"], axis=1)
```

```python
data1.head()
```

```python
data1['type'].value_counts()
```

```python
data1.dropna()
```

```python
 data1.drop_duplicates()
```

```python
data1.shape
```

```python
data2 = data1.fillna(2)
```

```python
data2.head()
```

```python
year_counts = data2['release_year'].value_counts().sort_index()
```

```python
import plotly.express as px

fig = px.line(
    x=year_counts.index,
    y=year_counts.values,
    title="Netflix Titles Released Over the Years",
    labels={'x': 'Release Year', 'y': 'Number of Titles'}
)
fig.show()
```

```python

genre_counts = data2['listed_in'].str.split(',').explode().value_counts()


plt.figure(figsize=(10,11))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette="viridis")
plt.title("Distribution of Netflix Titles by Genre")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

```python
country_counts = data2['country'].dropna().str.split(',').explode().value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=country_counts.values, y=country_counts.index, palette="coolwarm")
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

```python
type_counts = data2['type'].dropna().value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=type_counts.index, y=type_counts.values, palette="coolwarm")
plt.title("Total Content of Netflix" )
plt.xlabel("Type")
plt.ylabel("Count")  
plt.show()
```

```python
rating_counts = data2['rating'].value_counts()
percentages = rating_counts / rating_counts.sum() * 100
filtered_counts = rating_counts[percentages >= 0.9]
others_count = rating_counts[percentages < 0.9].sum()
if others_count > 0:
    filtered_counts = pd.concat([filtered_counts, pd.Series({'Others': others_count})])
plt.figure(figsize=(10, 10))
plt.pie(
    filtered_counts, 
    labels=filtered_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette("pastel", len(filtered_counts)),
    labeldistance=1.2
)
plt.title("Distribution of Content Ratings on Netflix", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(
    type_counts, 
    labels=type_counts.index, 
    autopct='%1.1f%%', 
    colors=sns.color_palette("pastel"), 
    startangle=90)
plt.title("Proportion of Movies and TV Shows on Netflix", fontsize=16)
plt.show()
```
